# Efecto de las características de adaptación de una neurona AdEx.

Primer caso: Varío el tiempo de adaptación y mantengo fijo el peso de la sinapsis.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy
import nest
nest.Install('scmodule')
nest.SetKernelStatus({"resolution": 0.01})
```

```python
num_neurons = 1 #número de neuronas que asigno a la capa del generador y a la de entrada
num_neu_sc = 3 
#número de neuronas para la capa SC (una para cada constante de tiempo de adaptación que voy a probar)
sigma = 8 #ms
```

Defino los parámetros del generador y creo una neurona con sus características.
Realizo el mismo paso con la capa de entrada (FEF).
En el caso de la capa SC creo una población de tres neuronas para poder evaluar las diferentes constantes de tiempo de adaptación en cada una de ellas.

```python
GEN_dict = {"i0":3.0, "beta":0.03, "gamma":1.8, "pop":0.5, "sc_onset":0.0, "distance":0.0}
GEN_pop = nest.Create("sc_generator", num_neurons, params=GEN_dict)
```

```python
FEF_dict = {"C_m":50.0,"t_ref":0.0,"V_reset":-55.0,"E_L":-70.0, "g_L":2.0, "I_e":0.0, "a":0.0, "b":60.0, 
            "Delta_T":2.0, "tau_w":30.0, "V_th":-50.0, "V_peak":-30.0, 'gsl_error_tol':1e-6}
FEF_pop = nest.Create("aeif_cond_exp", num_neurons, params=FEF_dict)
```

Los parámetros de la capa SC van a estar definidos por las constantes que se muestran a continuación junto con las diferentes variaciones de la constante de tiempo de adaptación.

```python
tau_w = [66.3, 44.8, 23.4]
SC_list = []
for tau in tau_w:
        SC_list.append({"C_m":280.0,"t_ref":0.0, "V_reset":-45.0, "E_L":-70.0, "g_L":10.0,"I_e":0.0, "a":4.0,
                        "b":80.0, "Delta_T":2.0,"V_th":-50.0, "V_peak":-30.0,"E_ex":0.0, "tau_syn_ex":5.0,
                        "E_in":-80.0, "tau_syn_in":10.0, 'gsl_error_tol':1e-6, "tau_w":tau})      
SC_pop = nest.Create("aeif_cond_exp",num_neu_sc, params=SC_list)
```

Realizo las conexiones, entre el generador y la capa de entrada, y entre la capa de entrada y la de SC.
La sinápsis entre las capas FEF y SC se define con un peso fijo de 13.0 nS.

```python
nest.Connect(GEN_pop, FEF_pop, "one_to_one", syn_spec={'weight':1.0})
nest.Connect(FEF_pop, SC_pop, "all_to_all",  syn_spec={'weight':13.0})
```

Creo y conecto un multímetro para grabar el comportamiento del potencial de membrana ('V_m') y la corriente de adaptación ('w'), y un detector de spikes para comprobar cuándo y en qué neurona se producen.

```python
multimeter=nest.Create("multimeter", params={"withtime":True, "record_from":["V_m", "w"], 'interval' :0.1})
nest.Connect(multimeter, SC_pop)

spikedetector = nest.Create("spike_detector", params={"withgid":True, "withtime":True})
nest.Connect(SC_pop, spikedetector)
```

```python
nest.Simulate(200.0)
```

```python
dmm=nest.GetStatus(multimeter)[0]
Vms=dmm["events"]["V_m"]
ts=dmm["events"]["times"]
neuronid = dmm["events"]["senders"]
q=dmm["events"]["w"]
q=q*1e-3

#Primera neurona de SC_pop
plt.figure()
plt.subplot(211)
plt.plot(ts[neuronid==SC_pop[0]], Vms[neuronid==SC_pop[0]], 'b')
plt.ylabel('Membrane potential [mV]')
plt.xlabel('Time (ms)')
plt.legend(['First neuron'])
plt.subplot(212)
plt.plot(ts[neuronid==SC_pop[0]], q[neuronid==SC_pop[0]], 'b')
plt.ylabel('Adaptative current q [nA]')
plt.xlabel('Time (ms)')


#Segunda neurona de SC_pop
plt.figure()
plt.subplot(211)
plt.plot(ts[neuronid==SC_pop[1]], Vms[neuronid==SC_pop[1]], 'g')
plt.ylabel('Membrane potential [mV]')
plt.xlabel('Time (ms)')
plt.legend(['Second neuron'])
plt.subplot(212)
plt.plot(ts[neuronid==SC_pop[1]], q[neuronid==SC_pop[1]], 'g')
plt.ylabel('Adaptative current q [nA]')
plt.xlabel('Time (ms)')

#Tercera neurona de SC_pop
plt.figure()
plt.subplot(211)
plt.plot(ts[neuronid==SC_pop[2]], Vms[neuronid==SC_pop[2]], 'r')
plt.ylabel('Membrane potential [mV]')
plt.xlabel('Time (ms)')
plt.legend(['Third neuron'])
plt.subplot(212)
plt.plot(ts[neuronid==SC_pop[2]], q[neuronid==SC_pop[2]], 'r')
plt.ylabel('Adaptative current q [nA]')
plt.xlabel('Time (ms)')
```

Fig.6. Evolución temporal de las variables de estado (potencial de membrana y corrientede adaptación), variando el tiempo de adaptación y manteniendo fijo el peso de la sinapsis.

```python
#spikedetector
dSD = nest.GetStatus(spikedetector)[0]
evs = dSD["events"]["senders"]
tspk = dSD["events"]["times"]
```

Utilizo de nuevo la gaussiana para calcular las densidades de los spikes.

```python
def gaussian_funct(time_diff, sigma):
    return (1/(sigma*numpy.sqrt(2*numpy.pi))*numpy.exp(-(time_diff*time_diff)/(2*(sigma*sigma))))
```

```python
t = numpy.arange(0.0,300.0,0.1)
gauss = numpy.zeros((len(SC_pop),t.shape[0]))
for idx,neurid in enumerate(SC_pop):    
    selected = (evs==neurid)
    for spike in tspk[selected]:        
        time_diff = t-spike
        gauss[idx,:] = gauss[idx,:]+(gaussian_funct(time_diff, sigma))
gauss = gauss*1e3
```

```python
plt.figure()
plt.xlim([0,150])
plt.plot(t, gauss.T)
plt.xlabel('Time (ms)')
plt.ylabel('Spike density (spk/s)')
plt.legend(['tau = 66.3 ms', 'tau = 44.8 ms', 'tau = 23.4 ms'])
```

Fig.7. Función de densidad de los spikes con pesos sinápticos fijos.
