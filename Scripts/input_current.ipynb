# Corriente de entrada, Iext, a las neuronas de la capa FEF.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy
import nest
```

Defino las variables.

```python
#For the logarithmic mapping function
Bu = 1.4 #mm
A = 3 #deg
r = 5 #deg
#For the input current
I0 = 3.0 #pA Scaling constant
gamma = 1.8 #Skewness and peak of the burst
beta = 0.03 #Measure for burst duration
pop = 0.5 #mm Recruited population size
```

Logarithmic mapping function, obtengo la posición de la neurona donde realizamos el sacádico.

```python
#no se utiliza en este apartado
ut = Bu*numpy.log((r+A)/A)
print(ut)
```

Cálculo de la corriente de entrada a las neuronas FEF.

```python
distance = numpy.linspace(0,0.825,12) #vector de distancias para una muestra de 12 neuronas
time = numpy.linspace(0.0,300.0,302) #vector de tiempos
first_part = I0*numpy.exp(-((distance*distance))/(2*(pop**2))) #primera parte de la ecuación Iext
second_part = numpy.power(time,gamma)*numpy.exp(-beta*time) #segunda parte
first_repeat = numpy.repeat([first_part],time.shape[0],axis=0) 
Iext = first_repeat*second_part[:,None]
```

```python
plt.figure()
plt.plot(time,Iext)
plt.ylabel('Current(pA)')
plt.xlabel('Time(ms)')
plt.legend(['0 mm', '0.06875 mm', '0.1375 mm', '0.20625 mm', '0.275 mm', '0.34375 mm', '0.4125 mm',
            '0.48125 mm', '0.55 mm', '0.61875 mm', '0.6875 mm','0.7625 mm', '0.825 mm']) 
#distancia de las neuronas con respecto a la neurona central donde se produce el sacádico
```
